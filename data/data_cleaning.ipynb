# Copy 
It's important to make a copy of the original dataframe before altering it too much. Just like when we alter a list. 


```python
df_filter = df.copy()
```

```python
df_filter['Artist(s)'] = df_filter['Artist(s)'].astype('string')
df_filter.dtypes
```

```python

df_filter['Title'] = df['Title'].str.replace(r'\(.*\)', '', regex=True)

df_filter
```

```python

df_filter['Release Date'] = df_filter['Release Date'].str.replace(r'\(.*\)', '', regex=True)

df_filter
```

```python

df_filter['Release Date'] = df_filter['Release Date'].str.replace(r'\[\d+\]', '', regex=True)

df_filter
```

```python

df_filter['Genres'] = df_filter['Genres'].astype('string')
df_filter.dtypes
```

```python


df_filter['Genres'] = df_filter['Genres'].str.replace(r'\[\d+\]', '', regex=True)
df_filter['Genres'] = df_filter['Genres'].str.replace(r'\\n', ', ', regex=True)


df_filter
```
